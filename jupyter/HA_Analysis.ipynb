```python

_heart_data_without_output = _heart_data.drop(columns=['output'])
_heart_data_without_output.head(5)
```

```python
from pca import pca
# Initialize and keep all PCs
model = pca()
# Fit transform
out = model.fit_transform(_heart_data_without_output)
```

```python
out['PC']
```

```python
model.scatter(label=True, legend=False)
```

```python
# The Biplot in Fig. 7 displays a different visual representation
# of the first two PCs. The axes of biplot represents the first two
# PCs. The rows of the eigenvector matrix is shown as a vector.
# Each of the observations in the dataset is drawn as a dot on
# the plot.
```

```python
A = out['loadings'].T
```

```python
sns.scatterplot(data=A, x="PC1", y="PC2")
plt.xlabel('')
plt.ylabel('')
for i in range(A.shape[0]):
 plt.text(x=A.PC1[i]+0.02,y=A.PC2[i]+0.02, s=_columns[i],
          fontdict=dict(color='red',size=10),
          bbox=dict(facecolor='yellow',alpha=0.5))
```

```python
Fig. 6 represents the PC coefficient plot. It visually represents
the amount of contribution each feature has on the first two
PCs. The figure supports the previous calculation of PCs and it
can be clearly seen from the figure that chol has
the highest contributions in the first PC. On the other hand, age, trtbps, oldpeak, restecg
and thalachh has the greatest contribution to the second PC. The
thalachh (location=(0,-0.98)) has the only negative coefficient
and is located bottom left side of the plot and far away from the clusters of features on the right side. 
```

```python
import numpy as np
VR = out['variance_ratio']
x = np.arange(len(VR)) + 1
plt.plot(x, VR, 'ro-', lw=3)
plt.xticks(x, [""+str(i) for i in x], rotation=0)
plt.xlabel('Number of components')
plt.ylabel('Explained variance') 
plt.show()
```

```python
data = _heart_data.sample(frac=0.9, random_state=786)
data_unseen = _heart_data.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
     
```
